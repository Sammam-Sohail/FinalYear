```python
import pandas as pd
import re
from csv import writer
```

```python
colnames=['Name','Date','Time', 'Text','Cleaned_text'] 
exchanges = ['BINANCE', 'BYBIT', 'FTX', 'HUOBI', 'KUCOIN', 'POLONIEX','KRAKEN']
df = pd.read_csv('./1_btctradingclub.csv',names = colnames,header=None)
print(df['Cleaned_text'][0])
print(df['Text'][0])
df.head()
```

```python
exchanges = 'Binance|kucoin|bitbuy'
conversion = 'usdt|usd|btc|eth'
exchangeregex = '|'.join(exchanges)
target = ('|').join(['1','2','3'])
# coins=''
def extractor(i):
  print(i)
  text = re.sub('[\',\[\]]', '', i)
  coin_name,conversion_name,call_exchange,call_type,call_duration,leverage,targets,stop_loss = ["None"]*8
  if(re.search('free\s*signal',text,re.IGNORECASE) and re.search('target|targets',text,re.IGNORECASE)):    

    coin_name=re.findall('(\w+)(?:usdt|usd|btc|eth)',text,re.IGNORECASE)[0]
    
    conversion_name = re.findall('\w+({})'.format(conversion),str(text),re.IGNORECASE)[0]
    
    """exchange not used"""
    call_exchange = re.findall('\w+({})'.format(exchanges),text,re.IGNORECASE)[0]

    call_type = re.findall('sell|buy',text,re.IGNORECASE)[0]

    """duration not used that much"""
    call_duration = re.findall('long|short',text,re.IGNORECASE)[0]

    """current not given around is given"""
    # current_value=re.findall('')
    
    leverage = re.findall('[ ]*[-]*[ ]*(\d+)x',text,re.IGNORECASE)

    # targets = re.findall("(\d*[.]?\d*)[ ]?(-)?[ ]?(\d*[.]?\d*)[ ]?-[ ]?(\d*[.]?\d*)[ ]?-[ ]?(\d*[.]?\d*)[ ]?-[ ]?(\d*[.]?\d*)",text,re.IGNORECASE)
    """some targets are not extracted although format is same """
    T=dict()
    targets=re.findall(r'targets\s*(\d.\d*)\s*-*\s*(\d.\d*)\s*-*\s*(\d.\d*)\s*-*\s*(\d.\d*)',text,re.IGNORECASE)
   

    """ stop loss used rarely also without value"""
    stop_loss = re.findall('stop\s*loss\s*-*\s*(\d.\d*)',text,re.IGNORECASE)[0]

  return (coin_name,conversion_name,call_exchange,call_type,call_duration,leverage,targets,stop_loss)
  
df["Call"] = df['Cleaned_text'].apply(extractor)
```
